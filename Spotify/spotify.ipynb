## Import chunked dataset

```python
import pandas as pd
df = pd.read_csv(r'E:\Data analytics\Spotify\dataset1.csv')
df.head()
df.info()
df.describe()
df.shape
df.head()
```

## Cleaning the dataset

## 1. Check for missing values

```python
df.isnull().sum()
```

## 2.Handle missing values

```python
df = df.dropna()
df.head()
```

## 3 Convert Data types

```python
df['track_id'] = df['track_id'].astype(str)
df['explicit'] = df['explicit'].astype(bool)
df.head()
```

## 4 Remove Duplicates

```python
df = df.drop_duplicates()
df.head()
```

## Scatter plot between 'popularity' and 'track_genre'

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Select only the first 10 rows
df_subset = df.head(10)

# Plotting a scatter plot between 'popularity' and 'track_genre'
sns.scatterplot(x='popularity', y='track_genre', data=df_subset)
plt.xlabel('Popularity')
plt.ylabel('Track Genre')
plt.title('Popularity vs Track Genre (First 10 rows)')
plt.show()
```

## Bar chart : Popularity vs Genre

```python
import matplotlib.pyplot as plt

# Select only the first 50 rows
df_subset = df.head(100)

# Plotting a bar chart between 'popularity' and 'track_genre'
plt.figure(figsize=(12, 8))
plt.bar(df_subset.index, df_subset['popularity'])
plt.xlabel('Index')
plt.ylabel('Popularity')
plt.title('Popularity vs Index (First 50 rows)')
plt.xticks(df_subset.index, df_subset['track_genre'], rotation=45, ha='right')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")

plt.figure(figsize=(12, 8))
plt.hist(df['popularity'], bins=50, edgecolor='black')
plt.xlabel('Popularity')
plt.ylabel('Frequency')
plt.title('Distribution of Popularity Values')
plt.show()
```

## Box plot for 'popularity' by 'explicit' status

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='explicit', y='popularity', data=df)
plt.xlabel('Explicit')
plt.ylabel('Popularity')
plt.title('Popularity by Explicit Status')
plt.show()
```

## Pair plot for numerical columns

```python
sns.pairplot(df[['popularity', 'danceability', 'energy', 'loudness', 'valence']])
plt.show()
```

## Bar chart: Mean Popularity for Each Key

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='key', y='popularity', data=df)
plt.xlabel('Key')
plt.ylabel('Mean Popularity')
plt.title('Bar Chart: Mean Popularity for Each Key')
plt.show()
```

## Scatter plot : Popularity vs Energy


```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")

plt.figure(figsize=(10, 6))
sns.scatterplot(x='energy', y='popularity', data=df)
plt.xlabel('Energy')
plt.ylabel('Popularity')
plt.title('Scatter Plot: Popularity vs. Energy')
plt.show()
```
